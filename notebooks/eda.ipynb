```python
import statsbomb as sb 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm import tqdm
```

```python
# read the DataFrame of competiton
comps = sb.Competitions().get_dataframe()
```

```python
comps
```

```python
# extract La Liga matches
competition_id = 11
season_ids = comps['season_id'][comps['competition_id']==competition_id].tolist()
season_names = comps['season_name'][comps['competition_id']==competition_id].tolist()
for i, season_id in enumerate(season_ids):
    matches_ = sb.Matches(event_id=competition_id, season_id=season_id).get_dataframe()
    # convert season id to season name 
    matches_['season'] = season_names[i]
    matches = matches_ if i == 0 else pd.concat([matches_, matches])
matches = matches.reset_index(drop=True)
```

```python
matches
```

```python
# read shot data
for i, event_id in tqdm(enumerate(matches['match_id'])):
    events = sb.Events(event_id=str(event_id))
    data_ = events.get_dataframe(event_type='shot')
    # add the 'season' column to the DataFrame
    data_.insert(0, 'season', matches['season'][i])
    # add the 'match_date' columns to the DataFrame
    data_.insert(1, 'match_date', matches['match_date'][i])
    data = data_ if i == 0 else pd.concat([data, data_])
data = data.reset_index(drop=True)
```

```python
# extract Barcelona's shot
shots = data[data['team']=='Barcelona'].reset_index(drop=True)
```

```python
# show the detail of data
print('total number of shots:', len(shots))
print()
 
print('columns -', list(shots))
print()

print('play_pattern -', (list(shots['play_pattern'].unique())))
print()

print('type -', list(shots['type'].unique()))
print()
 
print('technique -', list(shots['technique'].unique()))
print()

print('outcome -', list(shots['outcome'].unique()))
```

```python
# add new columns
# goal or no goal
shots['goal'] = shots['outcome'].apply(lambda x: 1 if x == 'Goal' else 0) 
# body_part: Left Foot & Right Foot -> Foot
shots['body_part'] = shots['body_part'].where((shots['body_part']!='Left Foot') | (shots['body_part']!='Right Foot'), 'Foot') # 左足・右足 - > 足
# distance from shot-taker to goal
shots['distance'] = np.sqrt(((120 - shots['start_location_x'])**2 + (40 - shots['start_location_y'])**2))
# angle between following two vectors: 
    # vector from shot-taker to right post
    # vector from shot-taker to left post
angles = []
for i in range(len(shots)):
    right_post = np.array([120 - shots['start_location_x'].iloc[i], 44 - shots['start_location_y'].iloc[i]])
    left_post = np.array([120 - shots['start_location_x'].iloc[i], 36 - shots['start_location_y'].iloc[i]])
    I = np.inner(right_post, left_post)
    R = np.linalg.norm(right_post)
    L = np.linalg.norm(left_post)
    angles.append(np.rad2deg(np.arccos(I / (R * L))))
shots['angle'] = angles
```

```python
# シュートデータから必要な列のみ抜き出す
cols = ['season', 'match_date', 'period', 'timestamp', 'team', 'player', 'play_pattern', 'under_pressure', 'body_part', 'type', 'technique', 'first_time', 'one_on_one', 'open_goal', 'key_pass_id', 'start_location_x', 'start_location_y', 'distance', 'angle', 'goal', 'statsbomb_xg']
shots = shots.loc[:, cols]
```

```python
def grouping(df, cols):
    """
    特定の列についてgroup化された特徴量を作成する
    
    Args:
        df (pd.DataFrame): 特徴量を含むdataframe
        cols (str or list): group by処理のkeyとなる列

    Returns:
        dfg (pd.DataFrame): 特定の列についてgroup化された特徴量群
    """
    dfg = df.groupby(cols).agg({'goal': ['count', 'mean']})

    return dfg

def plot_rel_col_goal(dfg, col, bar_width):
    """指定した列とゴールの関係を可視化する
    
    Args:
        dfg (pd.DataFrame): 特徴量を含むdataframe
        col (str): group by処理のkeyとなる列
        bar_widht(float): 棒グラフの棒幅
    """

    data = grouping(dfg, col)

    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()

    ax1.bar(data.index, data['goal']['count'], width=bar_width, label='Shot count', color='skyblue')
    ax2.plot(data.index, data['goal']['mean'], label='Conversion rate', color='orange', marker='.', markersize=10)

    h1, l1 = ax1.get_legend_handles_labels()
    h2, l2 = ax2.get_legend_handles_labels()
    ax1.legend(h1+h2, l1+l2, loc=2)

    ax1.set_title(col + " in J1 '16-'19", fontsize=14)
    ax1.set_ylim([0, data['goal']['count'].max() * 1.3])
    ax1.set_xticks(data.index)
    ax1.set_xticklabels(data.index,rotation=-90)
    ax1.set_ylabel('Shot count')

    ax2.set_ylim([0, data['goal']['mean'].max() * 1.3])
    ax2.set_ylabel('Conversion rate')
    
    plt.show()
```

```python
# 各変数ごとにゴール数とゴール率をプロット
# カテゴリカル変数
print("==============================")
print('categorical features')
print("==============================")
cat_cols = ['play_pattern', 'under_pressure', 'body_part', 'type', 'technique', 'first_time', 'one_on_one', 'open_goal']
for col in cat_cols:
    plot_rel_col_goal(train, col, 0.8)
    
# 連続変数（単位ごとにまとめる）
print("==============================")
print('numerical features')
print("==============================")
train_tmp = train.copy()
train_tmp['start_location_x'] = train_tmp['start_location_x'].astype('int') // 10 * 10
train_tmp['start_location_y'] = train_tmp['start_location_y'].astype('int') // 10 * 10
train_tmp['distance'] = train_tmp['distance'].astype('int') // 10 * 10
train_tmp['angle'] = train_tmp['angle'].astype('int') // 10 * 10
num_cols = ['start_location_x', 'start_location_y', 'distance', 'angle']
for col in num_cols:
    plot_rel_col_goal(train_tmp, col, 5)
```

```python
from matplotlib.patches import Ellipse

X_SIZE = 105.0
Y_SIZE = 68.0

BOX_HEIGHT = (16.5 * 2 + 7.32) / Y_SIZE * 100
BOX_WIDTH = 16.5 / X_SIZE * 100

GOAL = 7.32 / Y_SIZE * 100

GOAL_AREA_HEIGHT = 5.4864 * 2 / Y_SIZE * 100 + GOAL
GOAL_AREA_WIDTH = 5.4864 / X_SIZE * 100


def draw_pitch():
    """
    サッカーフィルドを描画するためのmatplotlib fig, axesオブジェクトを返す。
    
    Returns:
        fig (matplotlib.figure.Figure)
        axes (matplotlib.axes._subplots.AxesSubplot)
    """
        
    fig = plt.figure(figsize=(X_SIZE/15/2, Y_SIZE/15))
    axes = fig.add_subplot(1, 1, 1, facecolor='black')

    # X軸とY軸の目盛りを消す
    axes.xaxis.set_visible(False)
    axes.yaxis.set_visible(False)

    axes.set_xlim(0,100)
    axes.set_ylim(0,100)    

    axes = draw_patches(axes)
    
    return fig, axes

def draw_patches(axes):
    """
    ピッチ内のラインなどを描画するためのmatplotlib axesオブジェクトを返す。

    Returns:
        axes (matplotlib.axes._subplots.AxesSubplot)
    """

    plt.xlim([45,105])
    plt.ylim([-5,105])

    #　ピッチ
    axes.add_patch(plt.Rectangle((0, 0), 100, 100,
                                 edgecolor="white", facecolor="none", alpha=1))
                                 

    # ハーフウェーライン
    axes.add_line(plt.Line2D([50, 50], [100, 0],
                  c='w'))
    
    # センターサークル
    axes.add_patch(Ellipse((50, 50), 2*9.15/X_SIZE*100, 2*9.15/Y_SIZE*100,
                   ec='w', fc='none'))
    
    # ペナルティエリア
    axes.add_patch(plt.Rectangle((100-BOX_WIDTH, (100-BOX_HEIGHT)/2),  BOX_WIDTH, BOX_HEIGHT,
                   ec='w', fc='none'))
    axes.add_patch(plt.Rectangle((0, (100-BOX_HEIGHT)/2),  BOX_WIDTH, BOX_HEIGHT,
                   ec='w', fc='none'))                       
    
    # ゴールエリア
    axes.add_patch(plt.Rectangle((100-GOAL_AREA_WIDTH, (100-GOAL_AREA_HEIGHT)/2),  GOAL_AREA_WIDTH, GOAL_AREA_HEIGHT,
                   ec='w', fc='none'))
    axes.add_patch(plt.Rectangle((0, (100-GOAL_AREA_HEIGHT)/2),  GOAL_AREA_WIDTH, GOAL_AREA_HEIGHT,
                   ec='w', fc='none'))                       

    # ゴール
    axes.add_patch(plt.Rectangle((100, (100-GOAL)/2),  1, GOAL,
                   ec='w', fc='none'))
    axes.add_patch(plt.Rectangle((0, (100-GOAL)/2),  -1, GOAL,
                   ec='w', fc='none'))                       
    
    return axes

# References
# 1. https://github.com/rjtavares/football-crunching/blob/master/notebooks/working%20with%20positional%20data.ipynb
```

```python
from scipy.stats.kde import gaussian_kde

# シュートエリアを図示
x_mesh, y_mesh = np.mgrid[0:100:0.1, 0:100:0.1] # 10 cm単位でグリッド作成
loc_mesh = np.vstack([x_mesh.ravel(), y_mesh.ravel()])
x_shot = train['start_location_x'].values * 100 / 120 # X座標を補正
y_shot = (80 - train['start_location_y'].values) * 100 / 80 # Y座標を補正
loc_shot = np.vstack([x_shot, y_shot])
k_shot = gaussian_kde(loc_shot) # 2次元カーネル密度推定
z_shot = np.reshape(k_shot(loc_mesh).T, x_mesh.shape)

fig, ax = draw_pitch()
contour = ax.contourf(x_mesh, y_mesh, z_shot.reshape(x_mesh.shape), cmap='summer')
ax.set_title("Shot area in Barcelona's games Messi has played 04/05-17/18")
```

```python
# ゴールエリアを図示
x_goal = train['start_location_x'][train['goal']==1].values * 100 / 120 # X座標を補正
y_goal = (80 - train['start_location_y'][train['goal']==1].values) * 100 / 80 # Y座標を補正
loc_goal = np.vstack([x_goal, y_goal])
k_goal = gaussian_kde(loc_goal) # 2次元カーネル密度推定
z_goal = np.reshape(k_goal(loc_mesh).T, x_mesh.shape)

fig, ax = draw_pitch()
contour = ax.contourf(x_mesh, y_mesh, z_goal.reshape(x_mesh.shape), cmap='summer')
ax.set_title("Goal area in Barcelona's games Messi has played 04/05-17/18")
```
